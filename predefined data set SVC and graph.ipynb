# Problem Statement
<b> apply SVC model on predefined dataset of sklearn Brest Cancer and Wine Data and plot it on a graph with diffrent train size </b>
<p>Train Size : Test Size</br>
60 : 40</br>
70 : 30</br>
80 : 20</br>
</p>

```python
#importing pythone modules
from sklearn.datasets import load_breast_cancer
from sklearn.datasets import load_wine
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split 
import matplotlib.pyplot as plt
```

```python
#working on breast_cancer data
x=load_breast_cancer().data #decideing x data of breast cancer file
y=load_breast_cancer().target #decideing y data Targets of breast cancer file

list_test=[0.4,0.3,0.2] # creating list for accesing test_data_size


#creating model object
model_svc_cancer=SVC(kernel='linear')
accuracy_cancer=[]


#a loop for test and train data
for i in range(len(list_test)):
    xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=list_test[i],random_state=2)
    model_svc_cancer.fit(xtrain,ytrain)
    y_predict=model_svc_cancer.predict(xtrain)
    accuracy_cancer.append(model_svc_cancer.score(xtest,ytest)*100)
    
#printing accuracy list
print(accuracy_cancer)


#ploting graph

x_axis=['60% Train Data','70% Train Data','80% Train Data']
plt.figure(figsize=(10,10))
plt.style.use('ggplot')
bar_graph=plt.bar(x_axis,accuracy_cancer,label="SVC Model",width=.4)
plt.xlabel("Train Data Size in %",size=18)
plt.ylabel('Accuracy',Size=18)
plt.yticks(range(10,120,10))
plt.title("Accuracy Data On SVC Model on Diffrent Test Size on Breast Cancer Data",size=15)
plt.grid(True,linestyle='-.',c='orange')
plt.legend(loc='best',fontsize=13)


#creting function for auto label
def autolabel(rect):
    for i,text in enumerate (rect):
        plt.annotate('{0:0.2f}'.format(rect[i].get_height()),
                    xy=(rect[i].get_x(),rect[i].get_height()),
                    xytext=(37,4),
                    textcoords="offset points",
                    ha='center', va='bottom')
        
autolabel(bar_graph)
plt.show()
```
