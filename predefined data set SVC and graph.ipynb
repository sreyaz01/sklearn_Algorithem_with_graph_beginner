# Problem Statement
<b> apply SVC model on predefined dataset of sklearn Brest Cancer and Wine Data and plot it on a graph with diffrent train size </b>
<p>Train Size : Test Size</br>
60 : 40</br>
70 : 30</br>
80 : 20</br>
</p>

```python
#importing pythone modules
from sklearn.datasets import load_breast_cancer
from sklearn.datasets import load_wine
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split 
import matplotlib.pyplot as plt
import numpy as np
```

```python
#working on breast_cancer data
x=load_breast_cancer().data #decideing x data of breast cancer file
y=load_breast_cancer().target #decideing y data Targets of breast cancer file

list_test=[0.4,0.3,0.2] # creating list for accesing test_data_size


#creating model object
model_svc_cancer=SVC(kernel='linear')
model_rbf=SVC(gamma='auto')
accuracy_cancer=[]
accuracy_cancer_rbf=[]

#a loop for test and train data
for i in range(len(list_test)):
    xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=list_test[i],random_state=2)
    model_svc_cancer.fit(xtrain,ytrain)
    model_rbf.fit(xtrain,ytrain)
    y_predict=model_svc_cancer.predict(xtest)
    y_rbf_predict=model_rbf.predict(xtest)
    accuracy_cancer.append(model_svc_cancer.score(xtest,ytest)*100)
    accuracy_cancer_rbf.append(model_rbf.score(xtest,ytest)*100)
    
#printing accuracy list
print('Accuracy on Linear Kernel',accuracy_cancer)
print('Accuracy on RBF Kernel',accuracy_cancer_rbf)


#ploting graph

x_axis=['60% Train Data','70% Train Data','80% Train Data']
width=0.30 #bar width
x=np.arange(len(x_axis)) #the label locations


plt.figure(figsize=(10,10))
plt.style.use('ggplot')
ax=plt.subplot(111)

l_graph=ax.bar(x,accuracy_cancer,width,label="Linear Kernal")
rbf_graph=ax.bar(x+width,accuracy_cancer_rbf,width,label='RBF Kernel')
ax.set_xlabel("Train Data Size in %",size=18)
ax.set_ylabel('Accuracy',Size=18)
ax.set_yticks(range(10,120,10))
ax.set_xticks(x+width-0.18)
ax.set_xticklabels(x_axis)
ax.set_title("Accuracy Data On SVC Model on Diffrent Test Size on Breast Cancer Data",size=15)
ax.grid(True,linestyle='-.',c='orange')
ax.legend(loc='best',fontsize=13)


#creting function for auto label
def autolabel(rect):
    for i,text in enumerate (rect):
        plt.annotate('{0:0.2f}'.format(rect[i].get_height()),
                    xy=(rect[i].get_x(),rect[i].get_height()),
                    xytext=(23,4),
                    textcoords="offset points",
                    ha='center', va='bottom')
        
autolabel(l_graph)
autolabel(rbf_graph)
plt.show()
```
