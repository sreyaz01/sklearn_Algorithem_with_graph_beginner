```python
from sklearn import datasets #importing datasets which is implimented in sklearn
from sklearn.model_selection import train_test_split 
import numpy as np
from sklearn import linear_model #linear regression model import
```

```python
x=data.data #deciding x and y
y=data.target
```

```python
x
```

```python
y
```

```python
#spliting test and train data it will give 4 values so we need 4 variables to store 
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.30)

```

```python
xtest
```

```python
#creating model object
model=linear_model.LinearRegression()
```

```python
#train model
model.fit(xtrain,ytrain) #train machine according to data (find best fit line)
```

```python
#test model
p_values=model.predict(xtest) #predict values according to train data
```

```python
p_values #predicted values
```

```python
ytest #orignal ytest data
```

```python
#accuracy check
accuracy=model.score(xtest,ytest)*100
print(accuracy)
```

```python
import numpy as np
mean_square_error=np.mean((ytest-p_values)**2)
print(mean_square_error)
```
